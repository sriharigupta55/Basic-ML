```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x1 = np.array([0,0,1,1])
x2 = np.array([0,1,0,1])
y = np.array([0,1,1,0])
```

```python
data = []
for i in range(4):
    l = []
    l.append(np.exp(-((x1[i]-0)**2+(x2[i]-0)**2)**0.5))
    l.append(np.exp(-((x1[i]-1)**2+(x2[i]-1)**2)**0.5))
    data.append(l)
data = np.array(data)
plt.scatter(data[:,0],data[:,1],c=y)
```
